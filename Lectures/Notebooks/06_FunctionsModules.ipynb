## Functions

As our code becomes more complex, it makes sense to start breaking it up into units, called `functions` or `methods` that take data, perform operations on it, and possibly return one or more values.  They can be easily reused from anywhere in the program and allow for more readable and maintainable code. Constructing programs as a sequence of small logical functional units allows for code that is easier to write and debug.

```python
from IPython.display import Image
Image(filename='function.png')
```

```python
# let's define this function
def add(a,b):
    '''Add two numbers.'''
    c = a+b
    return c
```

```python
# The first comment line is used to describe the function when getting help.
help(add)
```

```python
# let's use it!
add(10,11)
```

Since python is dynamically typed, we can use other data types

```python
# floats
add(1.0,1.3)
```

```python
# strings
add('red','sox')
```

```python
# but we can't mix types
add('red',1)
```

```python
# since everything is a first class object in python, we can assign our function to a variable
func = add
func(10,10)
```

```python
# functions as reductions
reduce(add,range(1,11))
```

```python
# functions as maps
def cube(x): return x**3
cubes = map(cube,range(1,10))
print cubes
```

```python
# functions as filters
def odd(x): return (x%2)
odds = filter(odd,range(1,10))
print odds
```

### Lambdas

Python suports a very powerful construct borrowed from functional programming langues called lambdas.  They allow us to define functions on the fly.

```python
# using a function
def square(x): return x**2
map(square,range(5))
```

```python
# using a lambda, note this becomes one line
map(lambda x: x**2, range(5))
```

```python
# they can be used anywhere that a function is required
a = range(10)
a.sort(lambda x,y: cmp(y,x))
print a
```

## Modules

Don't reinvent the wheel! If you run into a programming problem, chances are, someone has already solved at least some part of it.  Modules are repositories of software and constants that we can `import` and use in our programs.

```python
# we import a module into our code using the `import` keyword
import math
```

```python
# use `dir` to get a list of methods contained in the module
dir(math)
```

```python
# use a method using a `.`
math.sqrt(1.345)
```

```python
# modules can also contain fundamental constants
print '%.16f' % math.pi
```

```python
# notice the contents of math preceeded and followed by two underscores
print math.__name__
```

### Working with times and dates

```python
import time
print time.strftime("%a, %d %b %Y   %H:%M:%S +0000",time.gmtime())
```

The module name acts as a namespace. This can become cumbersome, for very long module names and we can get around it by importing everything into the standard namespace.

```python
from time import *
print strftime("%m/%d/%Y %H:%M:%S",gmtime())
```

### Writing modules

Any code or collection of methods we write can be included in a module by writing them in a text editor (i.e. not in the iPython notebook or interpreter) and saving them with a unqiue filename `mymodule.py`.  We can treat these as complete python programs that can be run, or we can import them from other scripts to maximize effeciency and build collections of useful libraries.

Modules and scripts can both named `mymodule.py` and both can contain variables, functions/methods and classes.

** What's the difference?**

When imported, a module's `__name__` attribute is set to the module's file name, without *.py*.
When executed as a script, the `__name__` attribute is set to `__main__`.
Except for special cases, you shouldn't put any major executable code at the top-level. Put code in functions, classes, methods, and guard it with if `__name__ == __main__`.

```python
# filename: mymodule.py 

def foo():
    # some function
    return 1

# Main Function
def main():
    # put all your main program driver code here
    
# main is called once when the script is executed.    
if __name__ == "__main__":
    main()
```
